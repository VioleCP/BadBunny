Alguien me dijo que no le gustaba mucho Oasis de Bad Bunny porque le parecía que estaba lleno de resentimiento. Entonces, me propuse hacer una wordcloud para ver si los sentidos que más aparecen en las letras pueden ser asociados al rencor y resentimiento amoroso post ruptura. 

```python
import spacy
nlp = spacy.load('es_core_news_md')
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
stop_words = set(stopwords.words('spanish'))
import matplotlib.pyplot as plt
from wordcloud import WordCloud 

```

```python
#Importo un archivo txt que contiene todas las letras del disco
import os
with open ('data/OASIS.txt', 'r') as file:
    oasis = file.read()
print(oasis)
```

```python
#Tokenización y limpieza
oasis = oasis.lower()
```

```python
tokens = word_tokenize(oasis)
print(tokens)
```

```python
sinsw = [word for word in tokens if not word in stopwords.words("Spanish")] #Modifiqué desde la raíz la lista de stopwords para agregar algunas expresiones sonoras que se reiteraban en las canciones (como "ey")
limpio= [word for word in sinsw if word.isalnum()]
print(limpio)
```

```python
#Lemmatización
lem1= " ".join(limpio) 
print(lem1)
```

```python
doc=nlp(lem1)
```

```python
#Decidí solamente quedarme con los adjetivos y sustantivos porque me interesa ver cómo nombra al otro y cómo lo adjetiva
lem2 = [token.lemma_ for token in doc if token.pos_ == "ADJ" or token.pos_ == "NOUN"]
print(lem2)
```

```python
#WordCloud
wc = WordCloud(width=1300, height=1300, margin=0, max_words = 50, background_color='white').generate(" ".join(lem2))
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```

Incluso contemplando los límites de un análisis que tan solo remarca las palabras que más aparecen en el disco, resulta evidente que muchas de las palabras que aparecen hacen referencia al momento de ruptura (break, adiós) y tienen un deje resentido (odio, rencor).
